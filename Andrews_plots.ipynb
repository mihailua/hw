Розглянемо вибірку даних пацієнтів, що приймали 5 різних видів препаратів. Файл включає в себе дискретні спостереження як BP=NORMAL, BP=HIGH, дані спостереження замінено логічними відповідниками як 1, 0 та -1. Загалом, маємо три показника зо корелють із ефектом відповідних препаратів: Рівень тиску, Холестерол, Рівень Натрію до Кальцію.

Побудуємо діаграму Ендрюса для спостереження різниці дії між препаратами.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def andrews_curve(x,theta):
    curve = list()
    for th in theta:
        x1 = x[0] / np.sqrt(2)
        x2 = x[1] * np.sin(th)
        x3 = x[2] * np.cos(th)
        curve.append(x1+x2+x3)
    return curve

accuracy = 100
df = pd.read_csv('drug200.csv', delimiter=',')
theta = np.linspace(-np.pi, np.pi, accuracy)
for i in range(0,len(df.index)):
    colors = {'druga':'r', 'drugb':'g','drugc':'b','drugx':'y','drugy':'m'}
    sample = df.iloc[i,:].values.tolist()
    plt.plot(theta, andrews_curve(sample, theta),color=colors[sample[3].casefold()])
plt.show()
```

Як ми можемо побачити, чітко можна відрізнити лише 'DrugY', котрий на діаграмі позначили рожевим кольором.
